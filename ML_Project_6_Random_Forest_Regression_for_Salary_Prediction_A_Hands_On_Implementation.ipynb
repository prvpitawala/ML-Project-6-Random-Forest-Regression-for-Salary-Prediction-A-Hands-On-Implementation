In this notebook, we explore **Random Forest Regression**, a powerful ensemble learning technique used to predict continuous values. This approach leverages multiple decision trees to improve prediction accuracy and reduce the risk of overfitting.

# Random Forest Regression

**Random Forest Regression** is a type of ensemble learning method used for predicting continuous values. It builds multiple decision trees during training and combines their outputs to improve prediction accuracy and control overfitting.

## Importing the libraries

```
# importing part
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Position_Salaries.csv')
print(dataset)
```

```
# Divided to dependent variables and the independent variables
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```
print(X)
```

```
print(y)
```

## Training the Random Forest Regression model on the whole dataset

 training a Random Forest model with 10 trees.

```
from sklearn.ensemble import RandomForestRegressor

# Impliment random forest regression
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0) # n_estimators means trees in the forest

# Fit the model with X and y
regressor.fit(X, y)
```

## Predicting a new result

```
new_value = 6.5
regressor.predict([[new_value]])
```

## Visualising the Random Forest Regression results

```
# plot results
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (Random Forest Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
